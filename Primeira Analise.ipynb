```python
import pandas as pd
import matplotlib
```

```python
def ler_arquivo(ano):
    colunas = ['NU_ANO', 'CO_CURSO', 'NT_GER', 'NT_FG']
    df_nota = pd.read_csv(f'microdados_Enade_{ano}_LGPD\dados\microdados{ano}_arq3.txt',sep = ';',usecols = colunas )
    df_curso = pd.read_csv(f'microdados_Enade_{ano}_LGPD\dados\microdados{ano}_arq1.txt', sep = ';')
    
    # colocando valor 0 onde a nota for nula
    df_nota = df_nota.fillna(0)
    
    #como so precisamos relacionar cada CO_CURSO com suas respectivas caracteristicas, podemos remover as linhas duplicadas.
    df_curso = df_curso.drop_duplicates()
    
    # queremos somente as licenciaturas
    co_grupo = [702, 904, 905, 906, 1402, 1502, 1602, 2001, 2402, 2501, 3002, 3202, 3502, 4301, 5402, 6407]
    df_curso = df_curso[df_curso['CO_GRUPO'].isin(co_grupo)]
    
    #fizemos um inner join, somente mostrando os dados onde o CO_CURSO se encontra nos dois dataframes
    df_merged = df_nota.merge(df_curso, on = 'CO_CURSO')
    df_merged.drop('NU_ANO_y', axis = 1, inplace = True)
    
    # mapeando os estados
    mapeamento_estados = {
    11 : 'Rondônia (RO)',
    12 : 'Acre (AC)',
    13 : 'Amazonas (AM)',
    14 : 'Roraima (RR)',
    15 : 'Pará (PA)',
    16 : 'Amapa (AP)',
    17 : 'Tocantins (TO)',
    21 : 'Maranhão (MA)',
    22 : 'Piauí (PI)',
    23 : 'Ceará (CE)',
    24 : 'Rio Grande do Norte (RN)',
    25 : 'Paraíba (PB)',
    26 : 'Pernambuco (PE)',
    27 : 'Alagoas (AL)',
    28 : 'Sergipe (SE)',
    29 : 'Bahia (BA)',
    31 : 'Minas gerais (MG)',
    32 : 'Espírito Santo (ES)',
    33 : 'Rio de Janeiro (RJ)',
    35 : 'São Paulo (SP)',
    41 : 'Paraná (PR)',
    42 : 'Santa Catarina (SC)',
    43 : 'Rio Grande do Sul (RS)',
    50 : 'Mato Grosso do Sul (MS)',
    51 : 'Mato Grosso (MT)',
    52 : 'Goiás (GO)',
    53 : 'Distrito federal (DF)'}

    df_merged['CO_UF_CURSO'] = df_merged['CO_UF_CURSO'].map(mapeamento_estados)
    
    mapeamento_cursos = {
    702 : "Matemática (Licenciatura)",
    904 : "Letras-Português (Licenciatura)",
    905 : "Letras-Português e Inglês (Licenciatura)",
    906 : "Letras-Português e Espanhol (Licenciatura)",
    1402 : "Física (Licenciatura)",
    1502 : "Química (Licenciatura)",
    1602 : "Ciências Biológicas (Licenciatura)",
    2001 : "Pedagogia (Licenciatura)",
    2402 : "História (Licenciatura)",
    2501 : "Artes Visuais (Licenciatura)",
    3002 : "Geografia (Licenciatura)",
    3202 : "Filosofia (Licenciatura)",
    3502 : "Educação Física (Licenciatura)",
    4301 : "Música (Licenciatura)",
    5402 : "Ciências Sociais (Licenciatura)",
    6407 : "Letras-Inglês (Licenciatura)"}

    df_merged['CO_GRUPO'] = df_merged['CO_GRUPO'].map(mapeamento_cursos)
    
    return df_merged
```

```python
df_merged_2021 = ler_arquivo(2021)
```

### d) INDICADOR (% DE CONCLUINTES COM NG SUPERIOR A 75) E COUNT DE CONCLUINTES POR CURSO E POR ANO  (ANOS 2014, 2017 E 2021) (10)

```python
# agrupando por ano e curso e calculando o percentual com uma função lambda para 2014
df_agrupado_d_2014 = df_merged_2014.groupby('CO_GRUPO').agg(
    concluintes  =('NT_GER', 'count'),
    percentual_maior_75 = ('NT_GER', lambda x: (x > 75).mean() * 100)
).reset_index()

df_agrupado_d_2014['percentual_maior_75'] = df_agrupado_d_2014['percentual_maior_75'].round(2)


# agrupando por ano e curso e calculando o percentual com uma função lambda para 2017
df_agrupado_d_2017 = df_merged_2017.groupby('CO_GRUPO').agg(
    concluintes  =('NT_GER', 'count'),
    percentual_maior_75 = ('NT_GER', lambda x: (x > 75).mean() * 100)
).reset_index()

df_agrupado_d_2017['percentual_maior_75'] = df_agrupado_d_2017['percentual_maior_75'].round(2)


# agrupando por ano e curso e calculando o percentual com uma função lambda para 2021
df_agrupado_d_2021 = df_merged_2021.groupby('CO_GRUPO').agg(
    concluintes  =('NT_GER', 'count'),
    percentual_maior_75 = ('NT_GER', lambda x: (x > 75).mean() * 100)
).reset_index()
df_agrupado_d_2021['percentual_maior_75'] = df_agrupado_d_2021['percentual_maior_75'].round(2)
```

```python
"""
Explicando a função lambda. Ela sera aplicada para cada valor da coluna NT_GER.
Na parte (x > 75) sera feito uma verificação se o X (o valor da nota) é maior que 75. Isto ira retornar uma serie booleana,
onde cada valor é True se a nota for maior que 75 e False se for menor. O metodo .mean() calcula a media 
dos valores booleanos, onde ele ira considerar True como 1 e False como 0. Com isso, a media da serie booleana sera a 
proporçao de valores True, ou seja, a proporçao de quantos alunos tiraram mais que 75. O *100 é para converter em percentual.

Exemplo:

 Se temos estes valores para cada NT_GER [70, 80, 90]
 A serie boolena retornada pelo (x > 75) sera [False, True, True]. Calculando... (0 + 1 + 1) / 3 = 0.66
 Multiplicando por 100, 66% tiraram mais que 75.
 
 """
```
